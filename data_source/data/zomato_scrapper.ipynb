```python
import csv
from bs4 import BeautifulSoup
from selenium import webdriver
import requests
from urllib.request import Request, urlopen

from time import sleep
from random import randint

import mechanize
br = mechanize.Browser()
br.set_handle_robots(False)

#res = requests.get('https://ieeexplore.ieee.org/search/searchresult.jsp?newsearch=true&queryText=image%20processing')

#print(res.text)
#print(res.status_code)

url ='https://www.zomato.com/pune/baner-restaurants'
header={'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64; rv:78.0) Gecko/20100101 Firefox/78.0'}
res = requests.get(url,headers=header)

lenOfPage = browser.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
match=False
while(match==False):
    lastCount = lenOfPage
    time.sleep(5)
    lenOfPage = browser.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
    if lastCount==lenOfPage:
        match=True



def get_url(search_term):
    template = 'https://www.zomato.com/pune/{}'
    search_term = search_term.replace(' ', '-')
    url = template.format(search_term)
    
    
    return url

def extract_record(data):
    atag = data.find('a')
    try:
        url = 'https://www.zomato.com' + atag.get('href')
    except:
        url = " "
    try:
        name = data.find('p','sc-1hez2tp-0 sc-jBoNkH cqGniC').get_text()
    except:
        name = " "
    
    try:
        ratings = data.find('p','sc-1hez2tp-0 lhdg1m-2 dgzDPB').get_text()

    except:
        ratings = " "
    
    try:
        price_for_1 = data.find('p','sc-1hez2tp-0 sc-iIQDRA iplMES').get_text()

    except:
        price_for_1 = " "
    
    try:
        Delivery_Reviews = data.find('p','sc-1hez2tp-0 lhdg1m-8 hztxkg').get_text()

    except:
        Delivery_Reviews = " "
        
    try:
        off_given = data.find('p', 'sc-1hez2tp-0 sc-enOJk dUYOBQ').get_text() 

    except:
        off_given = " "
        
    try:
        Cuisine = data.find('p', 'sc-1hez2tp-0 sc-hfyccY ijjRYO').get_text()

    except:
        Cuisine = " "
    
    result = (url,name,ratings,price_for_1,Delivery_Reviews,off_given,Cuisine)
    
    return result

def main(search_term):
    browser = webdriver.Firefox(executable_path='/home/perman/Documents/geckodriver')

    
    records = []
    url = get_url(search_term)
    
    
    

    
    
    for page in range(1,3):
        driver.get(url.format(page))
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        results =soup.find_all('div',{"class":'jumbo-tracker'})
        sleep(randint(20,30))
        
        for data in results:
            record = extract_record(data)
            if record:
                records.append(record)
                
    driver.close()
    
    with open('results_zomato1.csv', 'w', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow(['url','name','ratings','price_for_1','Delivery_Reviews','off_given','Cuisine'])
        writer.writerows(records)
        
```

```python
import requests
from bs4 import BeautifulSoup
import pandas
import csv
import re

headers = {"User-Agent":"Mozilla/5.0 (platform; rv:geckoversion) Gecko/geckotrail Firefox/firefoxversion"}
page_no = 1
list_restaurants =[]

for page in range(1,3):
    print(page_no)
    url = "https://www.zomato.com/pune/restaurants?page={0}"
    driver.get(url.format(page))
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    search_list = soup.find_all("div", {'id': 'orig-search-list'})
    list_content = search_list[0].find_all("div", {'class': 'content'})
    for i in range(0,15):
        res_name = list_content[i].find("a", {'data-result-type': 'ResCard_Name'})
        locality = list_content[i].find("b")
        ratings = list_content[i].find("div", {'data-variation': 'mini inverted'})
        if ratings is None:
            continue
        rest6 = list_content[i].find_all("div", {'class': 'search-page-text clearfix row'})
        rest7 = rest6[0].find_all("span", {'class': 'col-s-11 col-m-12 nowrap pl0'})
        rest8 = rest7[0].find_all("a")
        cuisines = [e.string for e in rest8]
        cost_for_two = rest6[0].find("span", {'class': 'col-s-11 col-m-12 pl0'})
        if cost_for_two is None:
            continue
        votes = list_content[i].find("span", {'class': re.compile(r'rating-votes-div*')})
        if votes is None:
            continue
        dataframe ={}
        dataframe["rest_name"] = res_name.string.replace('\n', ' ')
        dataframe["locality"] = locality.string.replace('\n', ' ')
        dataframe["rating"] = ratings.string.replace('\n', ' ')
        dataframe["cuisines"] = cuisines
        dataframe["cost_for_two"] = cost_for_two.string[1:]
        dataframe["votes"] = votes.string.split()[0]
        list_restaurants.append(dataframe)
    page_no+=1
    
df = pandas.DataFrame(list_restaurants)
df.to_csv("zomato_aurants.csv")
```

```python
# Visit pyGuru on youTube


# importing modules
import requests
from bs4 import BeautifulSoup
import pandas as pd

# https://www.whoishostingthis.com/tools/user-agent/

city = input('Enter your city name : ')
url = 'https://www.zomato.com/'+city+'/top-restaurants'
header = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64; rv:78.0) Gecko/20100101 Firefox/78.0'}

response = requests.get(url,headers=header)
html = response.text

soup = BeautifulSoup(html,'lxml')
print(soup.title.text,'\n')

# scraping raw data and stroing it in a array

top_rest = soup.find_all('div',class_="bke1zw-0 cMipmx")
arr = []

for tr in top_rest:
	rsts = tr.find_all('div',class_="bke1zw-1")
	for i in rsts:
		name = i.select('a',class_='sc-muxYx sc-euitrJ hIVRVA')
		for k in name:
			arr.append(k.text)
	break

# print(arr)

# rearranging restaurant data in a single list

data = []
temp = [arr[0]]
for i in arr[1:]:
	if i.replace('.','',1).isdigit():
		data.append(temp)
		temp = [i]
	else:
		temp.append(i)

# print(data)

# strong different attributes in 4 lists

restraunts = []
address = []
cuisine = []
ratings = []

for lst in data:
	ratings.append(lst[0])
	restraunts.append(lst[1])
	address.append(lst[2])
	cuisine.append(','.join(lst[3:]))


header = ['Restraunt','Address','Cuisine','Ratings']
indices = [i for i in range(1,len(restraunts)+1)]
all_rests = zip(restraunts,address,cuisine,ratings)
df = pd.DataFrame(list(all_rests),index=indices,columns=header)
df.to_csv('Top restraunts in '+city+'.csv')
print(df)
```

```python
import scrapy
from scrapy.crawler import CrawlerProcess

class MarilynMansonQuotes(scrapy.Spider):
    name = "MarilynMansonQuotes"
    start_urls = [
        'https://en.wikiquote.org/wiki/Marilyn_Manson',
    ]
    
    def parse(self, response):
        for quote in response.css('div.mw-parser-output > ul > li'):
            yield {'quote': quote.extract()}

process = CrawlerProcess()
process.crawl(MarilynMansonQuotes)
process.start()
```
